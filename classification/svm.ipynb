```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import matplotlib
# fig=plt.figure(figsize=(1500, 1500), dpi= 150, facecolor='w', edgecolor='k')
matplotlib.rcParams['figure.figsize'] = [20, 10]
```

```python
data_loc = '../csv_files/Social_Network_Ads.csv'

file_data = pd.read_csv(data_loc)

X = file_data.iloc[:, [2, 3]].values
y = file_data.iloc[:, 4].values
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 100, random_state = 0)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(train_X, train_y)
y_pred = classifier.predict(test_X)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_y, y_pred)
cm
```

```python
def display(X_set, y_set):
    X_set, y_set = X_set, y_set
    X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                         np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
    plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
                 alpha = 0.75, cmap = ListedColormap(('red', 'green')))
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                    c = ListedColormap(('orange', 'green'))(i), label = j)
    plt.title('Logistic Regression (Test set)')
    plt.xlabel('Age')
    plt.ylabel('Estimated Salary')
    plt.legend()
    plt.show()
```
