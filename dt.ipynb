```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_loc = './csv_files/Position_Salaries.csv'
data_info = pd.read_csv(data_loc)


X = data_info.iloc[:, 1:2].values
y = data_info.iloc[:, 2].values
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
```

```python
regressor.fit(X, y)
```

```python
pred_y = regressor.predict([[121233.5]])
print(pred_y)

print(y)
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.xlabel('Position Label')
# plt.ylabel('Salary')
plt.show()
```

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regressor)')
plt.xlabel('Position Label')
plt.ylabel('Salary')
plt.show()
```
